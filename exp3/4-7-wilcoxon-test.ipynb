```python
import os
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import ks_2samp, kruskal, ttest_ind, t, kstwo, wilcoxon
from utils_best_arm import calc_return
from tqdm import tqdm
import numpy as np
import seaborn as sns
```

```python
path1 = "./ts-run-results/"
rewards = os.listdir("./Results/AAPL/exp3_concat/")
rewards = [reward.replace(".csv", "") for reward in rewards if reward not in ["volatility-10.csv", "sharpe-0.01.csv"]]

symbols = os.listdir("./Results/")[:]

indexes_df = []

ls = []
for symbol in symbols[:]:
    path1 = f"./Results/{symbol}/exp3_concat/{rewards[0]}.csv"
    df = pd.read_csv(path1).fillna(0)

    bh_percentage = df.iloc[:, 0].values
    test_portfolios = {0: df.iloc[:, 0].values}
    indexes = calc_return(bh_percentage, test_portfolios)
    ls.append(indexes)

bh_df = pd.concat(ls, axis=1).T
bh_df.columns = [f"B&H_{col.replace('_test', '')}" for col in bh_df.columns]
bh_df = pd.DataFrame(np.tile(bh_df, [100, 1]), columns=bh_df.columns)
bh_df.reset_index(drop=True, inplace=True)


for reward in rewards[:]:
    print(reward)
    
    ls2 = []
    for symbol in symbols[:]:
        path1 = f"./Results/{symbol}/exp3_concat/{reward}.csv"
        df = pd.read_csv(path1).fillna(0)

        bh_percentage = df.iloc[:, 0].values
        seeds = list(df.columns)[1:]
        ls = []
        for seed in tqdm(seeds):
            test_portfolios = {seed: df.loc[:, seed].values}
            indexes = calc_return(bh_percentage, test_portfolios)
            ls.append(indexes)

        ls_df = pd.concat(ls, axis=1).T
        ls_df.columns = [f"{reward}_{col.replace('_test', '')}" for col in ls_df.columns]
        ls2.append(ls_df)
    
    ls2 = pd.concat(ls2)
    if reward in ["MA", "MACD", "CCI", "MV"]:
        ls2 = pd.DataFrame(np.tile(ls2, [100, 1]), columns=ls2.columns)
    
    ls2 = ls2.reset_index(drop=True)
    indexes_df.append(ls2)


indexes_df = pd.concat(indexes_df, axis=1)
indexes_df.reset_index(drop=True, inplace=True)
indexes_df = pd.concat([bh_df, indexes_df], axis=1)
indexes_df.to_csv("./ts-run-results/[exp3]data-for-wilcoxon-test[2].csv", index=False)
indexes_df
```

```python
indexes_df = pd.read_csv("./ts-run-results/[exp3]data-for-wilcoxon-test.csv")
df_GMOE = pd.read_csv("./ts-run-results/[exp3]data-for-wilcoxon-test-GMOE.csv")

indexes_df2 = pd.concat([indexes_df, df_GMOE], axis=1)
indexes_df2
```

```python
criteria = ["sortino", "sharpe", "risk", "mdd", "downrisk", "cumreturn",]
test_df = []
for c in criteria:
    cols = [col for col in indexes_df2.columns if c == col.split("_")[-1]]

    tmp_df = indexes_df2[cols]

    n = tmp_df.shape[0] * 2 - 2

    vars = list(tmp_df.columns)

    for var1 in vars:
        if var1 in rewards: continue
        
        for _var in vars:
            if _var == var1:
                continue

            test = wilcoxon(tmp_df[var1], tmp_df[_var])

            _var1 = var1.split("_")[0]
            _var2 = _var.split("_")[-1]

            median1 = round(tmp_df[var1].median(), 4)
            median2 = round(tmp_df[_var].median(), 4)

            if _var2 in ["cumreturn", "sortino", "sharpe"]:
                if median1 > median2:
                    sign = 1
                else:
                    sign = 0
            else:
                if median1 < median2:
                    sign = 1
                else:
                    sign = 0
        

            test_df.append({
                "reward1": _var1,
                "var": _var2,
                "reward2": f"{_var.split('_')[0]}",
                "statistic": test.statistic,
                "pvalue": round(test.pvalue, 4),
                "median1": median1,
                "median2": median2,
                "median1 > median2": sign,
            })


test_df = pd.DataFrame(test_df)
test_df = test_df.sort_values(["reward1", "var"]).reset_index(drop=True)
test_df["var"] = test_df["var"].replace({
    "cumreturn": "CR",
    "downrisk": "DR",
    "mdd": "MDD",
    "risk": "Risk",
    "sharpe": "SR",
    "sortino": "STR",
})

dt = {
    "profit_0": "FP5",
    "profit-0": "FP5",

    "regularized_0.01": "FPR-0.01",
    "regularized_0.05": "FPR-0.05",
    "regularized_0.1": "FPR-0.1",
    "regularized_0.2": "FPR-0.2",

    "regularized-0.01": "FPR-0.01",
    "regularized-0.05": "FPR-0.05",
    "regularized-0.1": "FPR-0.1",
    "regularized-0.2": "FPR-0.2",

    "ts": "MSR-TSE",
    "greedy": "MSR-GME",
}

test_df["reward1"] = test_df["reward1"].replace(dt)
test_df["reward2"] = test_df["reward2"].replace(dt)


test_df.to_csv("./ts-run-results/[exp3]wilcoxon-test.csv", index=False)
test_df
```

```python
def convert_sig(x):
    if x < 0.01:
        return "***"
    elif 0.01 <= x < 0.05:
        return "**"
    elif 0.05 <= x < 0.1:
        return "*"
    else:
        return ""


def convert_star(x):
    if x == 1:
        return "√" 
    elif np.isnan(x):
        return "-"
    else:
        return ""


def construct_CR(var):
    print(var)
    tmp_df = test_df[test_df["var"] == var]
    CR_df1 = pd.pivot_table(tmp_df, values="statistic", index="reward1", columns="reward2").round(0).astype(str).applymap(lambda x: "-" if x == "nan" else x)
    CR_df1 = CR_df1.applymap(lambda x: x.replace(".0", ""))
    CR_df2 = pd.pivot_table(tmp_df, values="pvalue", index="reward1", columns="reward2").round(2)
    CR_df3 = CR_df2.applymap(convert_sig)
    CR_df2 = CR_df2.astype(str)
    CR_df4 = pd.pivot_table(tmp_df, values="median1 > median2", index="reward1", columns="reward2")
    CR_df4 = CR_df4.applymap(convert_star)
    # CR_df = CR_df1 + "^" + CR_df3 + "\n (" + CR_df2 + ", " +  CR_df4 + ")"
    # CR_df = CR_df.applymap(lambda x: "-" if "-^" in x else x)

    CR_df = CR_df1 + "^{" + CR_df3 + "}(" + CR_df4 + ")"
    CR_df = CR_df.applymap(lambda x: x.replace("()", ""))
    CR_df = CR_df.applymap(lambda x: "-" if "-^" in x else x)
    
    CR_df.index.name = "reward"
    columns = ["B&H", "CCI", "MA", "MACD", "FP5", "FPR-0.01", "FPR-0.05", "FPR-0.1", "FPR-0.2", "MSR-GME", "MSR-TSE", "GMOE"]
    CR_df = CR_df[columns].T[columns].T

    CR_df.to_csv(f"./ts-run-results/[exp3]wilcoxon-test-criteria-{var}.csv")

    return CR_df


vars = test_df["var"].unique().tolist()

for var in vars:
    construct_CR(var)
```
